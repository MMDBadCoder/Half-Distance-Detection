# تشخیص و اصلاح نیم فاصله ها و فاصله ها در متن و علائم نگارش

برای این پروژه یک دیتاست از لغات فارسی استفاده کرده ایم که شامل حدود بیست هزار لغت است و همچنین مجموعه‌ای از حروف اضافه که در نقش پیشوندو پسوند ظاهر می‌شوند را جمع‌آوری کرده‌ایم.
برای اصلاح مشکلات نگارشی و فاصله‌گذاری ها ابتدا متن را به مجموعه‌ای از لغات می‌شکنیم(به اصطلاح tokenisation) و سپس مجددا با اتصال لغات می‌پردازیم، و برای تشخیص شیوه‌ی صحیح فاصله‌گذاری از مجموعه داده‌های جمع‌آوری شده‌مان استفاده می‌کنیم.
ابتدا تمام متون را با کارکتر اسپیس از هم جدا می‌کنیم.
سپس هر کلمه را بررسی می‌کنیم که آیا اشکالات نگارشی دارد و یا آن را اصلاح می‌کنیم و یا آن را به مجموعه‌ای از لغات می‌شکنیم و لغت‌های کوچکتری می‌سازیم.
به طور مثال لغت «بعنوان» باید شکسته شود و به دو لغت «به» + «عنوان» تبدیل بشود. و ما برای این اصلاحات قائده‌ای در نظر گرفته‌ایم که در آن چک می‌کند آیا اگر حرف اول «ب» باشد و در صورت حذف آن باقی کلمه همچنان معنادار باشد، این تبدیل را انجام بدهد.
در مثال دیگر کلمه‌ی «آن‌چنان» را به «آنچنان» تبدیل می‌کند.
در مثالی دیگر کلمه‌ی «هم‌راه» را به «همراه» تغییر می‌دهد.
همچنین در مثالی دیگر همینکه را به «همینکه» تبدیل می‌کند.
حال ما تعداد زیادی لغت داریم که باید ببینیم با چه فاصله‌ای در کنار هم قرار می‌گیرند و برای این کار هر دو لغتی را با هم یکبار بررسی می‌کنیم. به طور مثال دو لغت «غم» و «انگیز» را کنار هم می‌گذاریم و چون لغت اول معنادار و لغت دوم بی‌معنی است و در لیست پسوند‌ها هست به هم متصل می‌شوند اما با یک نیم‌فاصله.
در مثالی دیگر دو لغت «جا» و «ها» را به هم وصل می‌کنیم.
همچنین حروف پیشوندی مثل «بی» را به لغات بعدی می‌چسبانیم.
